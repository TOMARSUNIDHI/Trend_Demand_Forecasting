```python
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.linear_model import Ridge, Lasso
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np

# Ensure x_train, y_train, x_test, and y_test are pandas DataFrames/Series
x_train = pd.DataFrame(x_train)
y_train = pd.Series(y_train)
x_test = pd.DataFrame(x_test)
y_test = pd.Series(y_test)

# Combine x_train and y_train into a single DataFrame for consistent cleaning
train_data = x_train.copy()
train_data['y'] = y_train

# Remove rows with missing values in either x_train or y_train
train_data_clean = train_data.dropna()

# Separate x_train and y_train after cleaning
x_train_clean = train_data_clean.drop(columns=['y'])
y_train_clean = train_data_clean['y']

# Impute missing values in x_train and x_test
imputer = SimpleImputer(strategy='mean')
x_train_imputed = imputer.fit_transform(x_train_clean)
x_test_imputed = imputer.transform(x_test)

# Ensure x_test and y_test have matching indices
x_test_clean = x_test.dropna()
y_test_clean = y_test.loc[x_test_clean.index]

# Fit the Ridge Regression model
rir = Ridge().fit(x_train_imputed, y_train_clean)
y_pred_rir = rir.predict(imputer.transform(x_test_clean))

# Evaluate the Ridge Regression model
MSE_rir = mean_squared_error(y_test_clean, y_pred_rir)
RMSE_rir = np.sqrt(MSE_rir)
MAE_rir = mean_absolute_error(y_test_clean, y_pred_rir)

print(f"Ridge Regression - MSE: {MSE_rir}, RMSE: {RMSE_rir}, MAE: {MAE_rir}")

# Fit the Lasso Regression model
lar = Lasso().fit(x_train_imputed, y_train_clean)
y_pred_lar = lar.predict(imputer.transform(x_test_clean))

# Evaluate the Lasso Regression model
MSE_lar = mean_squared_error(y_test_clean, y_pred_lar)
RMSE_lar = np.sqrt(MSE_lar)
MAE_lar = mean_absolute_error(y_test_clean, y_pred_lar)

print(f"Lasso Regression - MSE: {MSE_lar}, RMSE: {RMSE_lar}, MAE: {MAE_lar}")
```

```python
def get_vif(data):
    vif = pd.DataFrame()
    vif["Variables"] = data.columns
    vif["VIF"] = [ variance_inflation_factor(data.values,i) for i in range(data.shape[1])]
    return vif
```
